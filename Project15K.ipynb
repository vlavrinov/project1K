# **Финальный проект**. Telegram-бот для прогноза погоды



В этом проекте тебе нужно создать Telegram-бота, где пользователь сможет получить доступ к погодным данным через удобный интерфейс. Прогнозы на основе маршрута пользователя бот будет запрашивать у веб-сервиса, который создан в проектах 2 и 3 на основном уровне. Бот должен давать прогноз погоды на несколько дней вперёд для начальной, конечной и промежуточных точек маршрута.

**Цель:** используя `aiogram`, создать Telegram-бота, который будет выдавать прогноз погоды для разных временных интервалов и точек маршрута.

## Требования к функциональности проекта

1. **Обработка основных команд**
   - `/start` — приветствует пользователя и предоставляет краткое описание возможностей бота.
   - `/help` — отображает список доступных команд и выдаёт краткую инструкцию по использованию бота.
   - `/weather` — используется пользователем для запроса прогноза погоды. Команда запрашивает начальную и конечную точки маршрута, временной интервал прогноза и поддерживает добавление промежуточных остановок.

2. **Интеграция с веб-сервисом по прогнозу погоды**
   - Бот отправляет запросы к веб-сервису, используя параметры маршрута и временной интервал, которые задал пользователь.
   - Веб-сервис возвращает данные о погоде для заданных точек маршрута: температуру, осадки и ветер. Бот корректно обрабатывает эти данные и отображает их в удобном для пользователя формате.

3. **Поддержка интерактивных элементов**
   - В боте есть инлайн-кнопки и встроенные клавиатуры для удобного выбора временного интервала, подтверждения маршрута и управления другими параметрами.
   - Интерактивный интерфейс позволяет выбирать временной интервал прогноза (например, «Прогноз на 3 дня» или «Прогноз на 5 дней») и добавлять промежуточные точки маршрута.

4. **Обработка ошибок**
   - Бот корректно обрабатывает ошибки, например недоступность веб-сервиса или неверный формат данных.
   - Пользователь получает информативные сообщения об ошибках и рекомендации по корректному вводу данных.

5. **Вывод прогноза погоды:**
   - Прогноз отображается в понятном формате, с использованием текста или ссылок на визуализации, если веб-сервис их поддерживает.
   - Данные структурированы так, что пользователь легко понимает прогноз погоды для каждой точки маршрута.


### Критерии оценки

1. **Интеграция с веб-сервисом — 20 баллов**

   - Корректная работа с веб-сервисом (7 баллов):
     - бот отправляет корректные запросы к веб-сервису с параметрами маршрута (начальная, конечная и промежуточные точки) и временным интервалом прогноза;
     - бот обрабатывает полученные от веб-сервиса данные и выводит в удобном формате для пользователя.
   
   - Запросы для нескольких точек маршрута (7 баллов):
     - бот поддерживает запрос прогноза погоды для нескольких точек маршрута, включая промежуточные остановки;
     - прогнозы для каждой точки отображаются корректно и структурированно, пользователю легко в них ориентироваться.

   - Обработка ошибок веб-сервиса (6 баллов):
     - если веб-сервис недоступен или возвращает ошибку, бот корректно обрабатывает эту ситуацию и сообщает пользователю о проблеме;
     - предусмотрена обработка ошибок: недоступности API, неверных параметров запроса и тайм-аутов.

2. **Функциональность бота — 25 баллов**

   - Обработка команд (10 баллов):
     - команды `/start`, `/help`, и `/weather` реализованы корректно;
     - при запуске бот выводит краткое описание своих возможностей, по запросу — справку по командам.

   - Взаимодействие с пользователем через команду `/weather` (10 баллов):
     - бот запрашивает у пользователя маршрут (начальную и конечную точки, промежуточные остановки) и временной интервал прогноза;
     - пользователь выбирает временной интервал с помощью инлайн-кнопок или клавиатуры.

   - Отображение прогноза погоды (5 баллов):
     - прогнозы отображаются в удобном для чтения формате: текст, графики, ссылки на визуализации (если доступны);
     - пользователь может легко понять информацию о погоде: температуру, осадки и ветер.

3. **Интерактивные элементы и удобство использования — 20 баллов:**

   - Использование инлайн-кнопок и клавиатур (10 баллов):

     - добавлены поясняющие сообщения и, при необходимости, ссылки на визуализации, которые улучшают восприятие информации.
